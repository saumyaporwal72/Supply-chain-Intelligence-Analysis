**Step6: Fix Missing Values**

```python
print('=== MISSING VALUE AUDIT BEFORE CLEANING ===')
missing_before = df.isnull().sum()
missing_pct = (missing_before / len(df) * 100).round(1)
audit = pd.DataFrame({'Count': missing_before, 'Percent': missing_pct})
audit = audit[audit['Count'] > 0].sort_values('Count', ascending=False)
print(audit)
print()
```

```python
# --- STRATEGY 1: Drop columns with very high missing % ---
# Threshold: if more than 70% of a column is missing, it's not useful
```

```python
high_missing_cols = missing_pct[missing_pct > 70].index.tolist()
if high_missing_cols:
    print(f'Dropping high-missing columns: {high_missing_cols}')
    df.drop(columns=high_missing_cols, inplace=True)
else:
    print('No columns with >70% missing data found.')
```

```python
## # --- STRATEGY 2: Fill numeric missing values with median ---
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
filled_numeric = []
for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
        filled_numeric.append(f'{col}: filled with median {median_val:.2f}')
 
if filled_numeric:
    print('Numeric columns filled with median:')
    for msg in filled_numeric:
        print(f'  {msg}')
else:
    print('No numeric columns needed filling.')
```

```python
# --- STRATEGY 3: Fill categorical missing values with 'Unknown' ---
```

```python
cat_cols = df.select_dtypes(include=['object']).columns
filled_cat = []
for col in cat_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna('Unknown', inplace=True)
        filled_cat.append(col)
 
if filled_cat:
    print(f'Categorical columns filled with Unknown: {filled_cat}')
else:
    print('No categorical columns needed filling.')
```

```python
print()
print('=== MISSING VALUE AUDIT AFTER CLEANING ===')
remaining = df.isnull().sum().sum()
print(f'Total missing values remaining: {remaining}')
```

**Step8: Feature Engineering**

```python
print('Creating new features...')
```

```python
# --- FEATURE 1: Order Year and Month ---
# We need these to analyse monthly trends and seasonality
# Find the correct date column name after our cleaning,,
order_date_col = [c for c in df.columns if 'order_date' in c.lower()]
print(f'Order date column found: {order_date_col}')
```

```python
if order_date_col:
    col = order_date_col[0]
    df['order_year']       = df[col].dt.year
    df['order_month']      = df[col].dt.month
    df['order_month_name'] = df[col].dt.strftime('%b')   # Jan, Feb, Mar...
    df['order_quarter']    = df[col].dt.quarter           # Q1, Q2, Q3, Q4
    df['order_dayofweek']  = df[col].dt.day_name()       # Monday, Tuesday...
    print('  Created: order_year, order_month, order_month_name, order_quarter, order_dayofweek')
```

```python
# --- FEATURE 2: Delivery Gap ---
# How many days early (negative) or late (positive) was the delivery?
# Positive = late delivery, Negative = early, Zero = exactly on time
shipping_real = [c for c in df.columns if 'shipping_real' in c.lower() or 'days_for_shipping_real' in c.lower()]
shipping_sched = [c for c in df.columns if 'scheduled' in c.lower()]
if shipping_real and shipping_sched:
    df['delivery_gap_days'] = df[shipping_real[0]] - df[shipping_sched[0]]
    print('  Created: delivery_gap_days (positive = late, negative = early)')
```

```python
 
# --- FEATURE 3: Profit Margin Percentage ---
# Profit as a percentage of sales. This is more useful than raw profit.
# A $100 profit on a $200 sale (50% margin) is very different from
# a $100 profit on a $10,000 sale (1% margin)
if 'order_profit_per_order' in df.columns and 'sales' in df.columns:
    df['profit_margin_pct'] = (
        df['order_profit_per_order'] / df['sales'].replace(0, np.nan) * 100
    ).round(2)
    print('  Created: profit_margin_pct')
```

```python
 
# --- FEATURE 4: Binary Late Delivery Flag ---
# 1 = late, 0 = not late. Useful for aggregation and modelling.
if 'delivery_status' in df.columns:
    df['is_late_delivery'] = (
        df['delivery_status'].str.lower().str.contains('late').astype(int)
    )
    print('  Created: is_late_delivery (1=late, 0=on time)')
```

```python
# --- FEATURE 5: Revenue Tier ---
# Categorise orders into Small, Medium, Large, Enterprise by sales value
if 'sales' in df.columns:
    df['order_size_tier'] = pd.cut(
        df['sales'],
        bins=[0, 50, 200, 500, float('inf')],
        labels=['Small (<$50)', 'Medium ($50-200)', 'Large ($200-500)', 'Enterprise ($500+)'],
        include_lowest=True
    )
    print('  Created: order_size_tier')
```

```python
 
print()
print('All features created. New columns added to dataset:')
new_cols = ['order_year','order_month','order_month_name','order_quarter',
            'order_dayofweek','delivery_gap_days','profit_margin_pct',
            'is_late_delivery','order_size_tier']
for col in new_cols:
    if col in df.columns:
        print(f'  {col}: {df[col].dtype} | Sample: {df[col].iloc[0]}')
```

**Summary**

**Dataset:** DataCo Supply Chain Dataset, 180,519 rows, 53 columns

**Issues Found:**


- Column names had spaces, mixed cases, and special characters (fixed to snake_case)
- Date columns (order_date, shipping_date) were loaded as object/string type, not datetime
- [X] numeric columns had missing values (filled with column median)
- [X] categorical columns had missing values (filled with 'Unknown')
- [X] duplicate rows found and removed
- Sales and profit columns had right-skewed distributions
  (mean much higher than median, confirming outliers exist)





**Decisions Made:**
- Used median (not mean) to fill numeric missing values because sales data
  is right-skewed â mean is inflated by large outliers and would bias the fill
- Used 'Unknown' for categorical missing values because Unknown is meaningful
  information â it tells us data was not captured for that record
- Dropped columns with more than 70% missing values as they cannot
  contribute meaningful analysis
- Kept extreme profit margin values (very negative or very positive) because
  investigation confirmed they are real transactions, not data errors

**New Features Created:**
- delivery_gap_days: actual shipping days minus scheduled days
  (positive = late, negative = early, zero = on time)
- profit_margin_pct: order profit divided by sales Ã 100
  (normalises profit relative to order size for fair comparison)
- is_late_delivery: binary flag (1 = late, 0 = on time)
  (enables simple percentage calculations)
- order_year, order_month, order_month_name, order_quarter
  (extracted from order_date for time series analysis)
- order_size_tier: Small / Medium / Large / Enterprise
  (bucketed by sales value using pandas cut)

**Output:** cleaned_supply_chain.csv â 180519 rows Ã 60 columns,
zero missing values, all data types correct, ready for SQL and EDA analysis
